## 1. 데이터 로드 및 기본 정보

```python
import pandas as pd

FILE_PATH = 'data/training_data.csv' 

def perform_comprehensive_analysis(file_path):
    try:
        # 데이터 로드 및 메모리 로드
        print(f"데이터 로드 시작: {file_path}")
        df = pd.read_csv(file_path)
        print("\n------------------------------------------------------------------")
        
        # 1. 데이터 크기 및 타입
        print("1. 데이터 크기 및 타입")
        df.info()
        print("\n------------------------------------------------------------------")

        # 2. 샘플 확인 (상위 10개)
        print("2. 데이터 샘플 확인")
        print(df.head(10))
        print("\n------------------------------------------------------------------")
        
        # 3. 기본 통계 (문자열 데이터)
        print("3. 기본 통계 (문자열 데이터: title, author, subject)")
        # unique, top, freq 확인
        print(df.describe(include='object'))
        print("- unique: 고유값의 갯수")
        print("- top: 최빈값")
        print("- freq: 최빈값의 갯수")
        print("\n------------------------------------------------------------------")
        
        # 4. 결측값 확인
        print("4. 칼럼별 결측치 (Null/NaN) 개수 확인")
        print(df.isnull().sum())
        
    except FileNotFoundError:
        print(f"파일 경로 오류 {file_path}")
    except Exception as e:
        print(f"데이터 분석 중 오류 {e}")

if __name__ == '__main__':
    perform_comprehensive_analysis(FILE_PATH)
```

### 1. 데이터 크기 및 타입
- 행 수(RangeIndex): 28525
- 열 수(Data Columns): 3 (title, author, subject)
- memory: 668.7 KB
- data type: object

### 2. 데이터 샘플 확인
- 0~9 인덱스에 해당하는 도서의 title, author, subject가 잘 나오고 있는 것을 알 수 있음

### 3. 기본 통계
- count: 총 갯수 (Range Index와 일치) 
- unique: 고유값의 갯수 (책 제목 및 저자는 중복이 존재하는 것이 자연스러움)
- top: 최빈값 (gedichte: 독일어에서 '시', '이야기'의 복수형 명사) 
- freq: 최빈값의 빈도

### 4. 결측치 개수
- 없음 (양질의 데이터로 평가됨)

## 2. 시각화 및 분포 분석

```python
import matplotlib.pyplot as plt
import seaborn as sns
import platform
from matplotlib import font_manager, rc
import langid

FILE_PATH = 'data/training_data.csv'
plt.style.use('ggplot')
sns.set_palette('deep')

# 한글 폰트 설정: 경고 메시지 방지 및 콘솔 출력 안정화
if platform.system() == 'Windows':
    try:
        font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
        rc('font', family=font_name)
    except FileNotFoundError:
        pass # 폰트 설정 실패 시 무시
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False 

# 데이터 시각화
def perform_visualization_analysis(file_path):
    try:
        df = pd.read_csv(file_path)
        print(f"데이터 로드 완료. 총 {len(df):,}개 레코드.")
    except FileNotFoundError:
        print(f"파일 경로 오류: {file_path}")
        return
    
    total_records = len(df)

    # =========================================================================
    # 1. 언어 분포 분석 
    # =========================================================================
    print("\n------------------------------------------------------------------")
    print("1. Title 칼럼 언어 분포 분석 및 시각화")

    # Title 칼럼의 언어 감지 및 분류 (langid 사용)
    df['language'] = df['title'].apply(lambda x: langid.classify(str(x))[0])
    
    # 언어별 개수 및 비율 계산
    lang_counts = df['language'].value_counts()
    
    # 주요 언어 (상위 5개)만 시각화 
    top_5_langs = lang_counts.head(5)
    
    print("--- 언어 분포 통계 ---")
    for lang, count in lang_counts.items():
        percentage = (count / total_records) * 100
        print(f"- {lang.upper()} ({lang}): {count:,}개 ({percentage:.2f}%)")
    
    # 4. 시각화 (막대 그래프)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=top_5_langs.index, y=top_5_langs.values)
    
    plt.title('Top 5 Language Distribution of Title Column', fontsize=16)
    plt.xlabel('Language Code', fontsize=12)
    plt.ylabel('Number of Records', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show() 
    
    # =========================================================================
    # 2. subject 분포 (막대 그래프)
    # =========================================================================
    print("\n------------------------------------------------------------------")
    print("2. subject 분포 시각화 및 분석")
    
    subject_counts = df['subject'].value_counts()
    plt.figure(figsize=(10, 6))

    # 막대 그래프 생성
    sns.barplot(x=subject_counts.index, y=subject_counts.values)
    plt.title('Distribution of Records by Subject', fontsize=16)
    plt.xlabel('Subject', fontsize=12)
    plt.ylabel('Number of Records', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show() 

    print(f"총 레코드 수: {total_records:,}개")
    
    for subject, count in subject_counts.items():
        percentage = (count / total_records) * 100
        print(f"- {subject}: {count:,}개 ({percentage:.2f}%)")


    # =========================================================================
    # 3. 텍스트 길이 분포 (히스토그램)
    # =========================================================================
    print("\n------------------------------------------------------------------")
    print("3. title 길이 분포 분석")
    
    # 문자 길이 분포 확인
    df['title_len'] = df['title'].str.len()
    plt.figure(figsize=(10, 6))
    
    # Histogram 생성 - 문자 길이
    sns.histplot(df['title_len'], bins=50, kde=True)
    plt.title('Character Length Distribution of Title Column', fontsize=16)
    plt.xlabel('Character Length (Characters)', fontsize=12)
    plt.ylabel('Count of Books', fontsize=12)
    plt.axvline(df['title_len'].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {df["title_len"].mean():.2f}')
    plt.legend(title='Average Length') 
    plt.show()
    
    # 단어 개수 분포 확인
    df['title_word_count'] = df['title'].apply(lambda x: len(str(x).split()))
    plt.figure(figsize=(10, 6))
    
    # Histogram 생성 - 단어 수
    sns.histplot(df['title_word_count'], bins=30, kde=True)
    plt.title('Word Count Distribution of Title Column', fontsize=16)
    plt.xlabel('Word Count (Words)', fontsize=12)
    plt.ylabel('Count of Books', fontsize=12)
    plt.axvline(df['title_word_count'].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {df["title_word_count"].mean():.2f}')
    plt.legend(title='Average Count')
    plt.show()
    
    
    print("\n분석- 데이터 전처리 방향 설정")
    
    # 주요 통계 출력
    len_desc = df['title_len'].describe(percentiles=[.9, .95, .99])
    word_desc = df['title_word_count'].describe(percentiles=[.9, .95, .99])
    
    print("1. 문자 길이 주요 통계:")
    print(len_desc)
    print(f"평균 길이: {len_desc['mean']:.2f} 문자")
    print(f"최대 길이: {len_desc['max']:.0f} 문자")
    print(f"99% 레코드 커버 길이: {len_desc['99%']:.0f} 문자")
    
    print("\n2. 단어 수 주요 통계:")
    print(word_desc)
    print(f"평균 단어 수: {word_desc['mean']:.2f} 단어")
    print(f"최대 단어 수: {word_desc['max']:.0f} 단어")
    print(f"99% 레코드 커버 단어 수: {word_desc['99%']:.0f} 단어")

if __name__ == '__main__':
    perform_visualization_analysis(FILE_PATH)
```
