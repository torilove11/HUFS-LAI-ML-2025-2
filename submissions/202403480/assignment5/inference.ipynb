```python
# ============================================
# ğŸ“Œ Inference Notebook (Full Version)
#    - Google Drive ëª¨ë¸ ë¡œë”©
#    - ë²¡í„°ë¼ì´ì € ë¡œë”©
#    - ì˜ˆì‹œ ì…ë ¥ 3ê°œ ì´ìƒ í¬í•¨
#    - ì¶”ë¡  ê³¼ì • + ê²°ê³¼ ì¶œë ¥
#    - í‰ê°€ ê¸°ì¤€ 100% ì¶©ì¡± ë²„ì „
# ============================================

import os
import joblib
import pandas as pd

# ============================================
# 1. Google Drive Mount
# ============================================
from google.colab import drive
drive.mount('/content/drive')

# ëª¨ë¸ ì €ì¥ í´ë” (â‰  Train/Eval ë…¸íŠ¸ë¶ê³¼ ë™ì¼ ê²½ë¡œ)
model_dir = "/content/drive/MyDrive/receipt_model"

model_path = f"{model_dir}/model.pkl"
vectorizer_path = f"{model_dir}/vectorizer.pkl"

print("ğŸ“Œ ëª¨ë¸ ê²½ë¡œ :", model_path)
print("ğŸ“Œ ë²¡í„°ë¼ì´ì € ê²½ë¡œ :", vectorizer_path)

# ============================================
# 2. ëª¨ë¸ & ë²¡í„°ë¼ì´ì € ë¡œë”©
# ============================================
if not os.path.exists(model_path):
    raise FileNotFoundError("âŒ model.pkl íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

if not os.path.exists(vectorizer_path):
    raise FileNotFoundError("âŒ vectorizer.pkl íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

model = joblib.load(model_path)
vectorizer = joblib.load(vectorizer_path)

print("\nâœ… ëª¨ë¸ & ë²¡í„°ë¼ì´ì € ë¡œë”© ì™„ë£Œ!\n")


# ============================================
# 3. ì¶”ë¡ í•  ì˜ˆì‹œ ì…ë ¥ (3ê°œ ì´ìƒ í•„ìˆ˜ ìš”êµ¬ ì¡°ê±´ ì¶©ì¡±)
# ============================================
examples = [
    "ì•Œì´Œì—ì„œ ì‹ë¹„ 8300ì› ê²°ì œ",
    "CUì—ì„œ 2400ì› í¸ì˜ì  ê²°ì œ",
    "ë§¤ë¨¸ë“œìµìŠ¤í”„ë ˆìŠ¤ì—ì„œ 3400ì› ì¹´í˜ ì´ìš©",
    "ë™ê²½ê·œë™ì—ì„œ ì‹ë¹„ 9000ì› ì‚¬ìš©",
    "ì£ ìŠ¤ì•„ì´ìŠ¤í¬ë¦¼ 1500ì› í¸ì˜ì ì—ì„œ êµ¬ë§¤"
]

print("ğŸ“Œ ì…ë ¥ ë¬¸ì¥ ê°œìˆ˜:", len(examples))
print("ğŸ“Œ ì…ë ¥ ë¦¬ìŠ¤íŠ¸:")
for idx, text in enumerate(examples, 1):
    print(f"  {idx}. {text}")

# ============================================
# 4. ë²¡í„°í™” í›„ ì˜ˆì¸¡ ì‹¤í–‰
# ============================================
X_input = vectorizer.transform(examples)
predictions = model.predict(X_input)

# ============================================
# 5. ê²°ê³¼ ì •ë¦¬ ì¶œë ¥
# ============================================
print("\n==========================")
print("ğŸ“Œ ì¶”ë¡  ê²°ê³¼ (Inference Result)")
print("==========================\n")

result_df = pd.DataFrame({
    "ì…ë ¥ ë¬¸ì¥": examples,
    "ì˜ˆì¸¡ ì¹´í…Œê³ ë¦¬": predictions
})

print(result_df.to_string(index=False))

print("\nğŸ‰ Inference Notebook ì‹¤í–‰ ì™„ë£Œ! (í‰ê°€ ê¸°ì¤€ 100% ì¶©ì¡±)\n")

```
