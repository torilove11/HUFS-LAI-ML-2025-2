```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#데이터 로드
df = pd.read_csv("/content/data.csv")

#데이터 크기 (행, 열)
print("Data Shape (행, 열):")
print(df.shape)
print()

#컬럼 이름 확인
print("Columns:")
print(df.columns.tolist())
print()

#각 컬럼별 데이터 타입
print("Data Types:")
print(df.dtypes)
print()

#처음 5행 미리보기
print("Head (top 5 rows):")
print(df.head())
print()

#기본 통계 정보
print("Basic Statistics (include='all'):")
print(df.describe(include='all'))
print()

#결측값 확인
print("Missing Values (각 컬럼별 NaN 개수):")
print(df.isnull().sum())
print()


#라벨 간 교차분포
print("Cross-tab: Mood × Tense")
display(pd.crosstab(df["mood"], df["tense"]))

print("\nCross-tab: Person × Tense")
display(pd.crosstab(df["person"], df["tense"]))

df["verb_len"] = df["verb"].astype(str).str.len()

print("'verb_len' 미리보기:")
print(df[["verb", "verb_len"]].head())
print()


# mood / tense / person 분포 (막대그래프)

cat_cols = ["mood", "tense", "person"]
for col in cat_cols:
    plt.figure(figsize=(6,4))
    df[col].value_counts().plot(kind="bar")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.grid(alpha=0.3)
    plt.show()

for col in ["mood", "tense", "person"]:
    df[col + "_id"] = df[col].astype("category").cat.codes

num_df = df[["verb_len", "mood_id", "tense_id", "person_id"]]
corr = num_df.corr()

print("Correlation matrix:")
print(corr)
print()


#verb 길이 분포 (히스토그램)
plt.figure(figsize=(6,4))
plt.hist(df["verb_len"], bins=20)
plt.title("Histogram of Verb Length")
plt.xlabel("verb_len (문자 수)")
plt.ylabel("Count")
plt.grid(alpha=0.3)
plt.show()

#상관계수 히트맵 (heatmap)
fig, ax = plt.subplots(figsize=(5,4))
cax = ax.imshow(corr, interpolation='nearest')
ax.set_title("Correlation Heatmap")

#축 라벨 설정
ax.set_xticks(range(len(num_df.columns)))
ax.set_yticks(range(len(num_df.columns)))
ax.set_xticklabels(num_df.columns, rotation=45, ha="right")
ax.set_yticklabels(num_df.columns)

#색상바 추가
fig.colorbar(cax)
plt.tight_layout()
plt.show()

#산점도
plt.figure(figsize=(6,4))
plt.scatter(df["mood_id"], df["verb_len"], alpha=0.25, color="orange")
plt.title("Scatter: mood_id vs verb_len")
plt.xlabel("mood_id (법 코드)")
plt.ylabel("verb_len (문자 수)")
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["person_id"], df["verb_len"], alpha=0.3)
plt.title("Scatter: person_id vs verb_len")
plt.xlabel("person_id (인칭 코드)")
plt.ylabel("verb_len (문자 수)")
plt.grid(alpha=0.3)
plt.show()

#코드 작성에 ChatGPT의 도움을 받았습니다
```
